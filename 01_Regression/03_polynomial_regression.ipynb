```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
## for demonastration we create a quadratic dataset

X= 6*np.random.rand(100,1)-3
y=0.5*X**2+1.5*X+2+np.random.randn(100,1)

plt.scatter(X,y,color="g")
plt.xlabel("Random X Dataset")
plt.ylabel("Y dataset")
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,
                                                 test_size=0.25,
                                                 random_state=42,
                                                 )
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Implement Simple linear Regression

from sklearn.linear_model import LinearRegression
regression_1= LinearRegression()
```

```python
regression_1.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score

r2score= r2_score(y_test,regression_1.predict(X_test))
print(r2score)
```

```python
## Visualise the regression
plt.plot(X_train,regression_1.predict(X_train),color='r')
plt.scatter(X_train,y_train)
plt.xlabel("X Dataset")
plt.ylabel("Y data")
plt.title(r"$h(x)=\beta_0+\beta_1x_1$")
```

```python
## Now lets do this using polynomial Regression

## first lets apply polynomial transformation. To increase degree, Suppose if
## we have one feature and we need to increase the features we can apply
## polynomial transform
from sklearn.preprocessing import PolynomialFeatures
```

```python

## lets use degree =2
poly = PolynomialFeatures(degree=2,include_bias=True)
X_train_poly=poly.fit_transform(X_train)
X_test_poly=poly.transform(X_test)
```

```python
## Now we have created the Polynomial daaset

X_train_poly
```

```python
# Implement Polynomial Simple linear Regression Using transformed datset

regression_2= LinearRegression()
regression_2.fit(X_train_poly,y_train)
r2score_poly= r2_score(y_test,regression_2.predict(X_test_poly))
print("Regression model 1 --- ",r2score,
      "\nCoefficient Model 1 --- ",regression_1.coef_,
      "\nIntercept Model 1 --- ",regression_1.intercept_)
print("\nRegression model 2 --- ",r2score_poly,
      "\nCoefficient Model 2 --- ",regression_2.coef_,
        "\nIntercept Model 2 --- ",regression_2.intercept_)
```

```python
plt.scatter(X_train,regression_2.predict(X_train_poly),color="r")
plt.scatter(X_train,y_train)
plt.xlabel("X Dataset")
plt.ylabel("Y Dataset")
```

```python
## lets use degree = 3
poly = PolynomialFeatures(degree=3,include_bias=True)
X_train_poly_3=poly.fit_transform(X_train)
X_test_poly_3=poly.transform(X_test)

X_train_poly_3
```

```python

regression_3= LinearRegression()
regression_3.fit(X_train_poly_3,y_train)
r2score_poly_3= r2_score(y_test,regression_3.predict(X_test_poly_3))
print("Regression model 1 --- ",r2score,
      "\nCoefficient Model 1 --- ",regression_1.coef_,
      "\nIntercept Model 1 --- ",regression_1.intercept_)
print("\nRegression model 2 --- ",r2score_poly,
      "\nCoefficient Model 2 --- ",regression_2.coef_,
        "\nIntercept Model 2 --- ",regression_2.intercept_)
print("\nRegression model 3 --- ",r2score_poly_3,
      "\nCoefficient Model 3 --- ",regression_3.coef_,
        "\nIntercept Model 3 --- ",regression_3.intercept_)
```

```python
plt.scatter(X_train,regression_3.predict(X_train_poly_3),color="r")
plt.scatter(X_train,y_train)
plt.xlabel("X Dataset")
plt.ylabel("Y Dataset")
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
```

```python
def poly_regression(degree):
    X_new = np.linspace(-3,3,200).reshape(200,1)
    poly_features = PolynomialFeatures(degree=degree,include_bias=True)
    lin_reg = LinearRegression()
    poly_regression = Pipeline([
        ("polynomial_features",poly_features),
        ("linear_regression",lin_reg)
    ])
    poly_regression.fit(X_train,y_train)
    y_pred_new = poly_regression.predict(X_new)

    ## ploting
    plt.plot(X_new, y_pred_new,'r',label='Degree'+str(degree),linewidth=3)
    plt.plot(X_train,y_train,'b.',linewidth=3)
    plt.plot(X_test,y_test,'g.',linewidth=3)
    plt.xlabel("X")
    plt.ylabel('y')
    plt.axis([-4,4,0,10])
    plt.show()
    
```

```python
poly_regression(2)
```
