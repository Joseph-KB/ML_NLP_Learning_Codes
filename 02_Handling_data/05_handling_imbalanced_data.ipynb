```python
import numpy as np
import pandas as pd

## whenever selecting data it should be same.
np.random.seed(123)

n_samples = 1000
class_0_ratio = 0.9
n_class_0 = int(n_samples*class_0_ratio)
n_class_1 = int(n_samples-n_class_0)
```

```python
## create imbalanced dataset
class_0 = pd.DataFrame({
    'feature_1':np.random.normal(loc=0,scale=1,size=n_class_0),
    'feature_2':np.random.normal(loc=0,scale=1,size=n_class_0),
    'target':[0]*n_class_0
})

class_1 = pd.DataFrame({
    'feature_1':np.random.normal(loc=0,scale=1,size=n_class_1),
    'feature_2':np.random.normal(loc=0,scale=1,size=n_class_1),
    'target':[1]*n_class_1
})
```

```python
class_0
```

```python
df = pd.concat([class_0,class_1]).reset_index(drop=True)
df
```

```python
df['target'].value_counts()
```

## Up Sampling

```python
df_minority = df[df['target']==1]
df_majority = df[df['target']==0]
```

```python
from sklearn.utils import resample
df_minority_upscaled = resample(df_minority,replace=True,
         n_samples=len(df_majority),
         random_state=42) 
# Upscaleing and Sample With Replacement


print(df_minority_upscaled.shape)
print(df_minority_upscaled.target.value_counts())
```

```python
df_upsampled = pd.concat([df_majority,df_minority_upscaled])
df_upsampled.target.value_counts()
```

## Downsample

```python
import numpy as np
import pandas as pd

## whenever selecting data it should be same.
np.random.seed(123)

n_samples = 1000
class_0_ratio = 0.9
n_class_0 = int(n_samples*class_0_ratio)
n_class_1 = int(n_samples-n_class_0)

## create imbalanced dataset
class_0 = pd.DataFrame({
    'feature_1':np.random.normal(loc=0,scale=1,size=n_class_0),
    'feature_2':np.random.normal(loc=0,scale=1,size=n_class_0),
    'target':[0]*n_class_0
})

class_1 = pd.DataFrame({
    'feature_1':np.random.normal(loc=0,scale=1,size=n_class_1),
    'feature_2':np.random.normal(loc=0,scale=1,size=n_class_1),
    'target':[1]*n_class_1
})

df = pd.concat([class_0,class_1]).reset_index(drop=True)

df['target'].value_counts()
```

```python
df_minority = df[df['target']==1]
df_majority = df[df['target']==0]
```

```python
## Downsampling
from sklearn.utils import resample
df_majority_downsample = resample(
    df_majority, replace=False,
    n_samples=len(df_minority),
    random_state=42
)
```

```python
df_minority.target.value_counts()
```

```python
df_majority_downsample.target.value_counts()
```

```python
df_downsampled = pd.concat([df_minority,df_majority_downsample])
```

```python
df_downsampled.reset_index(drop=True,inplace=True)
```
